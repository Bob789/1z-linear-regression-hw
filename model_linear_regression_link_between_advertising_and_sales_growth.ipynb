```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the input data: advertising investment reshaped as a column vector
advertising_investment = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50]).reshape(-1, 1)

# Define the target data: corresponding sales growth
sales_growth = np.array([25, 30, 40, 45, 50, 60, 65, 70, 80])

# Create a LinearRegression model object
model = LinearRegression()

# Fit the model to the data (learn the best-fit line)
model.fit(advertising_investment, sales_growth)

# Print the slope (coefficient of x)
print(f"Slope (m): {model.coef_[0]:.2f}")

# Print the intercept (where the line crosses the y-axis)
print(f"Intercept (b): {model.intercept_:.2f}")

# Format the equation of the regression line as a string
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")

# Predict sales growth for investment of 60
advertising_investment_to_predict = 60

# Define the Sales we want to predict: Y = (mx) + b
sales_growth_to_predict = (advertising_investment_to_predict * model.coef_[0]) + model.intercept_

print(f"To predict sales growth by investment of {advertising_investment_to_predict}, approximately {sales_growth_to_predict:.2f} growth is expected")

# Plotting section
plt.figure(figsize=(6, 4))  # Set the figure size

# Plot the original data points as blue dots
plt.scatter(advertising_investment, sales_growth, color='blue', label='Data points', s=60)

# Plot the regression line (predicted sales growth)
# Create a range of x values for smooth line
x_line = np.linspace(advertising_investment.min(), 65, 100).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression line', linewidth=2)

# Mark the prediction point (for investment = 60) in green
plt.scatter([advertising_investment_to_predict], [sales_growth_to_predict], 
           color='green', s=150, label=f'Prediction (Investment={advertising_investment_to_predict})', 
           marker='*', edgecolors='darkgreen', linewidth=2)

# Add chart title and axis labels
plt.title('Linear Regression - Advertising Investment vs. Sales Growth', fontsize=14, fontweight='bold')
plt.xlabel('Advertising Investment', fontsize=12)
plt.ylabel('Sales Growth', fontsize=12)
plt.grid(True, alpha=0.3)  # Add grid to the background
plt.legend(fontsize=10)    # Show legend

# Display the equation of the line on the plot
plt.text(15, 75, equation, fontsize=12, bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

# Calculate R-squared value
r2 = model.score(advertising_investment, sales_growth)
plt.text(12, 64, f"$R^2$ = {r2:.3f}", fontsize=12,
         bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

# Show the final plot
plt.tight_layout()
plt.show()
```
